## **Categorical Dependence - Challenge 2**



```
df_income = pd.read_csv('https://raw.githubusercontent.com/mervatkheir/CSEN1095-Data-Engineering/master/data/income_data.csv')
```

```
df_income.head() #inspect the dataframe
```

```
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
contigency= pd.crosstab(df_income['race'], df_income['education']) 
contigency
```

```
c,p,dof,e = chi2_contingency(contigency)
print(p)
```

Yes, there is a correlation between the race and education. This is indicated through the p value which is less than 0.05
